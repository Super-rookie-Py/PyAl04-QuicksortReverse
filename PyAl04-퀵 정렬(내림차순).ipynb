#### 피벗값보다 큰값과 작은값을 비교할 때 왼쪽부터 피벗 값 보다 큰값을 찾고
#### 오른쪽부터 피벗 보다 작은 값을 찾아서 두 값을 바꿔나가다가
#### 두 값의 위치가 엇갈릴 때 새로운 피벗값으로 두 부분을 나누어 값을 구해나간다.
#### 왼쪽부터 피벗값보다 작은값을, 오른쪽부터 피벗보다 큰값을 찾게끔 하면 내림차순으로 구현가능( 부등호 식만 바꾸면)

```python

data = [3, 7, 8, 1, 5, 9, 6, 10, 2, 4]
n = len(data)

def quicksort(data, start, end):
    
    piv = start             #피벗값 지정
    i = start + 1
    j = end
    
    if len(data) > 1: #원소가 1개인 경우
        while i <= j :
            
            while data[i] > data[piv] and i <len(data)-1:    #피벗값보다 큰 값을 찾는다.
                i = i + 1
                
            while data[j] < data[piv] and j > start:    #피벗값보다 작은 값을 찾는다. 이 두 부분의 부호만 바꿔주면 내림차순이된다.
                j = j - 1
            if i >= j:                                   # i값과 j 값이 엇갈릴 경우 피벗값과 값j(피벗보다 작은값)을 바꿔주고 재귀함수.
                print("세번째p: ", data[piv])
                data[j], data[piv] = data[piv], data[j]
                print("네번째p:", data[piv])
                print("마지막과정:", data)
                print("i: ",i, "j: ", j)
                print("piv",data[piv], "pipindex :", piv)
                quicksort(data, start, j-1)
                quicksort(data, j+1, n-1)
            else:
                print("첫번째I: ", data[i])
                print("두번째j: ", data[j])
                data[i], data[j] = data[j], data[i]               
                print("중간과정: ", data)
                print("i: ",i, "j: ", j)
                print("piv",data[piv], "pipindex :", piv)
                
    else:
        return
        
```

```python
quicksort(data, 0, n-1)
```
